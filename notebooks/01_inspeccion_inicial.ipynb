## 8. Análisis de valores faltantes

La presencia de valores faltantes es uno de los aspectos más importantes durante la inspección inicial de un conjunto de datos.

En esta sección se identificará la cantidad de valores nulos presentes en cada variable y se calculará el porcentaje que representan respecto del total de registros.

Esta información permitirá evaluar la magnitud del problema y servirá como evidencia para justificar las decisiones de limpieza y preparación de datos en etapas posteriores.

```python
# Cantidad de valores faltantes por columna
valores_faltantes = df.isnull().sum()

# Porcentaje de valores faltantes
porcentaje_faltantes = (valores_faltantes / len(df)) * 100

# Crear una tabla resumen
faltantes = pd.DataFrame({
    "Valores faltantes": valores_faltantes,
    "Porcentaje (%)": porcentaje_faltantes.round(2)
})

faltantes
```

### Interpretación

El análisis de valores faltantes permitió identificar que únicamente tres variables presentan datos ausentes: **monthly_watch_time_mins**, **favorite_genre** y **last_login_date**.

La variable con mayor cantidad de valores faltantes es **last_login_date**, con **320 registros** (3,92 %), seguida por **favorite_genre**, con **240 registros** (2,94 %), y **monthly_watch_time_mins**, con **193 registros** (2,37 %).

Las demás variables no presentan valores faltantes, por lo que contienen información completa para la totalidad de los registros.

En general, el porcentaje de datos faltantes observado es relativamente bajo. Sin embargo, antes de decidir si corresponde eliminar, imputar o conservar dichos registros, será necesario analizar el contexto de cada variable durante la etapa de preparación y limpieza de datos.

## 11. Análisis de las variables categóricas

En esta sección se analizarán las variables categóricas del conjunto de datos.

El objetivo es identificar las diferentes categorías presentes en cada variable y detectar posibles inconsistencias en la escritura, diferencias de formato o valores atípicos que puedan afectar el análisis posterior.

Esta inspección permitirá justificar futuras decisiones de normalización durante la etapa de preparación de datos.

```python
# Variables categóricas
variables_categoricas = [
    "subscription_plan",
    "country",
    "favorite_genre"
]

# Mostrar los valores únicos de cada variable
for variable in variables_categoricas:

    print("="*60)
    print(f"Variable: {variable}")
    print("="*60)

    print(df[variable].value_counts(dropna=False))

    print("\n")
```

### Interpretación

El análisis de las variables categóricas permitió identificar diversas inconsistencias en la forma en que fueron registrados los datos.

En la variable **subscription_plan** se observaron diferencias de escritura asociadas al uso de mayúsculas y minúsculas, ausencia de acentos, abreviaturas, traducciones al inglés y errores tipográficos.

La variable **country** presenta distintas formas de representar un mismo país, incluyendo nombres en español, inglés, abreviaturas y diferencias de capitalización.

Por su parte, la variable **favorite_genre** evidencia inconsistencias similares, además de traducciones entre español e inglés y diferencias ortográficas entre categorías equivalentes.

Estas observaciones indican que será necesario realizar un proceso de normalización durante la etapa de preparación de datos con el objetivo de unificar categorías equivalentes y garantizar la consistencia del análisis.
