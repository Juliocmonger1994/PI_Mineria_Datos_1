# 1. Introducción

En esta última etapa del Proyecto Integrador se presentan las conclusiones obtenidas a partir del proceso completo de minería de datos desarrollado sobre el conjunto de datos de usuarios de una plataforma de streaming.

A lo largo del proyecto se aplicó una metodología secuencial que incluyó la inspección inicial del dataset, la limpieza y preparación de los datos, el análisis exploratorio (EDA) y la aplicación del Análisis de Componentes Principales (PCA).

El objetivo de esta notebook es sintetizar los principales resultados alcanzados, evaluar el cumplimiento de los objetivos planteados, identificar las limitaciones del estudio y proponer posibles líneas de trabajo futuro.

Las conclusiones se fundamentan exclusivamente en la evidencia obtenida durante el desarrollo del proyecto, procurando distinguir claramente entre los resultados observados y las interpretaciones derivadas del análisis.

# 2. Síntesis del proceso de minería de datos

El desarrollo del proyecto se estructuró en cuatro etapas principales:

- **Inspección inicial del dataset**, con el propósito de comprender la estructura de los datos, identificar el tipo de variables y detectar problemas de calidad.
- **Limpieza y preparación de los datos**, corrigiendo registros duplicados, valores inconsistentes, categorías equivalentes, valores faltantes y tipos de datos inadecuados.
- **Análisis Exploratorio de Datos (EDA)**, mediante estadísticas descriptivas y visualizaciones que permitieron comprender la distribución de las variables y detectar patrones relevantes.
- **Análisis de Componentes Principales (PCA)**, aplicando previamente un proceso de estandarización para estudiar la estructura interna de las variables numéricas y evaluar la posibilidad de reducir la dimensionalidad.

Cada una de estas etapas permitió mejorar progresivamente la calidad del conjunto de datos y obtener información confiable para el análisis posterior.

# 3. Respuesta a los objetivos del proyecto

Los objetivos planteados al inicio del proyecto fueron alcanzados satisfactoriamente.

En primer lugar, se logró mejorar significativamente la calidad del conjunto de datos mediante un proceso sistemático de limpieza y preparación, eliminando inconsistencias, normalizando categorías e imputando valores faltantes cuando resultó metodológicamente adecuado.

Posteriormente, el análisis exploratorio permitió caracterizar el comportamiento de las principales variables, identificar distribuciones, valores atípicos y relaciones entre los distintos atributos del conjunto de datos.

Finalmente, la aplicación del Análisis de Componentes Principales permitió estudiar la estructura de las variables numéricas y concluir que la información se encuentra distribuida entre las tres componentes principales, por lo que no fue posible realizar una reducción significativa de la dimensionalidad sin pérdida de información.

# 4. Principales hallazgos

Entre los principales resultados obtenidos durante el proyecto pueden destacarse los siguientes:

- La etapa de limpieza permitió obtener un conjunto de datos consistente, eliminando registros duplicados y corrigiendo valores inconsistentes en variables numéricas y categóricas.
- La normalización de las variables categóricas permitió unificar categorías equivalentes, evitando interpretaciones erróneas durante el análisis exploratorio.
- El análisis exploratorio evidenció distribuciones asimétricas y la presencia de valores atípicos reales en algunas variables, los cuales fueron conservados cuando representaban comportamientos plausibles.
- El PCA mostró que las variables **age**, **monthly_watch_time_mins** y **customer_support_tickets** aportan información complementaria y que ninguna de ellas domina completamente la estructura del conjunto de datos.
- Como consecuencia, fue necesario conservar las tres componentes principales para representar la totalidad de la variabilidad del conjunto de datos.

# 5. Conclusiones generales

El proceso de minería de datos permitió transformar un conjunto de datos originalmente heterogéneo en una fuente de información organizada, consistente y adecuada para posteriores análisis.

La combinación de técnicas de preparación de datos, análisis exploratorio y reducción de dimensionalidad permitió comprender con mayor profundidad el comportamiento de las variables y la estructura del conjunto de datos.

Los resultados obtenidos ponen de manifiesto la importancia de realizar una adecuada etapa de limpieza antes de aplicar técnicas analíticas más avanzadas, ya que la calidad de los datos condiciona directamente la confiabilidad de las conclusiones obtenidas.

Asimismo, el análisis mediante PCA evidenció que, para este conjunto de datos, las variables numéricas describen dimensiones complementarias del comportamiento de los usuarios. En consecuencia, el PCA resultó especialmente útil como herramienta de exploración e interpretación, aunque no permitió reducir significativamente la dimensionalidad sin pérdida de información.

En conjunto, el proyecto cumplió los objetivos propuestos y generó un conjunto de datos preparado para futuras aplicaciones de aprendizaje automático y análisis predictivo.

# 6. Limitaciones

Las conclusiones alcanzadas se encuentran condicionadas por la calidad y el alcance de la información disponible en el conjunto de datos analizado.

Si bien durante la etapa de preparación se corrigieron inconsistencias, se normalizaron variables y se imputaron determinados valores faltantes, algunas decisiones metodológicas —como la conservación de registros sin fecha de último acceso (`last_login_date`)— responden a la ausencia de criterios objetivos para su reconstrucción.

Por otra parte, el análisis se centró en técnicas descriptivas y de reducción de dimensionalidad, por lo que no se desarrollaron modelos predictivos ni algoritmos supervisados que permitieran evaluar el poder predictivo de las variables disponibles.
